# Respuesta P3: Entrenamiento Modelos
Estamos ahora listos para entrenar los modelos. Los siguientes modelos serán entrenados y evaluados:
    1. Logistic Regression with Scikit-Learn
    2. Gradient boosting with XgBoost
    3. Neural Network with TensorFlow

```python
# import libraries
import pandas as pd
import numpy as np
import math
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score, confusion_matrix, hamming_loss, log_loss
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set();
```

```python
# leer data
consolidado = pd.read_csv('data/consolidado.csv')
```

Eliminaremos la variable `Unnamed: 0` y setearmos la variable `id_new` cómo index.

```python
# eliminar Unnamed:0
consolidado = consolidado.drop(columns=['Unnamed: 0'])
# set id_new como index
consolidado = consolidado.set_index('id_new')
# seleccionar variables con correlación mayor a 0.1
consolidado = consolidado[['danceability', 'energy', 'speechiness', 'instrumentalness', 'valence', 'song_type']]
```

```python
# separamos predictores y target y los convertimos a arrays para
# posteriormente balancear data
X, y = consolidado.drop('song_type', axis=1), consolidado['song_type']
```

Ahora procederemos a balancear la data usando subsampling y oversammpling

```python
# balancear data
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=0)
X_resampled, y_resampled = ros.fit_sample(X, y)
# convertir a pandas dataframe
X_resampled = pd.DataFrame(X_resampled)
X_resampled.columns = list(consolidado.drop('song_type', axis=1).columns.values)
y_resampled = pd.DataFrame(y_resampled)
y_resampled.columns = ['song_type']
consolidado_resampled = pd.concat([X_resampled, y_resampled], axis=1)
```

```python
# tabla de frecuencias para target
pd.crosstab(index=y_resampled["song_type"],  
                              columns="count") 
```

```python
# visualizar categorías balanceadas vs distribución de predictores
colnames = consolidado_resampled.columns
rows = [0,0,0,1,1,1]
cols = [0,1,2,0,1,2]
f, axes = plt.subplots(2, 3, figsize=(20, 12), sharex=False)
for colname,i,j in list(zip(colnames,rows,cols)):
    sns.boxplot(x="song_type", y=colname, data=consolidado_resampled,
             ax=axes[i, j]);
```

```python
# pairplots variables con correlation mayor a 0.1 (checkear PART 1)
sns.pairplot(consolidado_resampled, hue = 'song_type');
```

```python
# generar training and test datasets
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size = 0.3)
```

```python
# transformar to numpy array y reshape y vectors
X_train, X_test, y_train, y_test = np.asarray(X_train), np.asarray(X_test), np.asarray(y_train), np.asarray(y_test) 
y_train = y_train.reshape(-1)
y_test = y_test.reshape(-1)
```

### 1. Logistic Regression

```python
# import libraries
from sklearn import linear_model
```

```python
# define model
lreg = linear_model.LogisticRegression(C=1e5)
```

```python
# fit data
lreg.fit(X_train, y_train)
```

```python
# predecir en el test set
lreg_pred = lreg.predict(X_test)
```

```python
# calcular medidas de error
lreg_ac = accuracy_score(y_test, lreg_pred)
lreg_hl = hamming_loss(y_test, lreg_pred)
lreg_ll = log_loss(y_test, lreg_pred)
print("Accuracy Score: %.2f"%lreg_ac)
print("Hamming Loss: %.2f"%lreg_hl)
print("Log Loss: %.2f"%lreg_ll)
```

```python
# matriz de confusion
confusion_matrix(y_test, lreg_pred)
```

### 2. Gradient Boosting

```python
# import libraries
import xgboost
```

```python
# XGBoost hyperparamters tunning
n_estimators = [50, 100, 150, 200, 250]
max_depth = [2, 4, 6, 8, 10]
gamma = [5, 10, 15, 20, 25]
results = []
for i in n_estimators:
    for j in max_depth:
        for k in gamma:
            
            xgb = xgboost.XGBClassifier(n_estimators=i, learning_rate=0.05, 
                                       gamma=k, subsample=0.75,colsample_bytree=1, max_depth=j)
            xgb.fit(X_train,y_train)
            xgb_pred = xgb.predict(X_test)
            xgb_acc = accuracy_score(y_test, lreg_pred)
            results.append([i,j,k,xgb_acc])
```

```python
# encontrar la mejor combinación de parámetros
results = pd.DataFrame(results)
results.columns = ['n_estimators', 'max_depth', 'gamma', 'acc_score']
results.iloc[results['acc_score'].idxmax()]
```

```python
# entrenamos el modelo con la mejor combinación de parámetros encontrada
xgb = xgboost.XGBClassifier(n_estimators=50, learning_rate=0.05, 
                                       gamma=5, subsample=0.75,colsample_bytree=1, max_depth=2)
```

```python
# fit model a la data
xgb.fit(X_train,y_train)
```

```python
# predecir en el dataset de test
xgb_pred = xgb.predict(X_test)
```

```python
# calcular medidas de error
xgb_ac = accuracy_score(y_test, xgb_pred)
xgb_hl = hamming_loss(y_test, xgb_pred)
xgb_ll = log_loss(y_test, xgb_pred)
print("Accuracy Score: %.2f"%xgb_ac)
print("Hamming Loss: %.2f"%xgb_hl)
print("Log Loss: %.2f"%xgb_ll)
```

```python
# matriz de confusion
confusion_matrix(y_test, xgb_pred)
```

### 3. Neural Network

```python
# import labraries
import tensorflow as tf
```

```python
# definir predictores
feat_cols_dict = {}
for i in X.columns:
     feat_cols_dict["{0}".format(i)] = tf.feature_column.numeric_column("{0}".format(i))
feat_cols = [feat_cols_dict[x] for x in feat_cols_dict.keys()]
feat_cols
```

```python
# crear modelo
# transformar train sets a pandas dataframe
X_train = pd.DataFrame(X_train)
X_train.columns = list(consolidado.drop('song_type', axis=1).columns.values)
y_train = pd.DataFrame(y_train)
y_train.columns = ['song_type']
consolidado_resampled = pd.concat([X_train, y_train], axis=1)[[
    'danceability', 'energy', 'speechiness', 'instrumentalness',
    'valence', 'song_type']]
input_func = tf.estimator.inputs.pandas_input_fn(x=X_train, y=y_train, batch_size=40, num_epochs=10000, shuffle=True)
model = tf.estimator.DNNClassifier(feature_columns=feat_cols, hidden_units=[10,10,10])
```

```python
# entrenar modelo
model.train(input_fn=input_func,steps=2000)
```

```python
# predecir en test set
# transformar train sets a pandas dataframe
X_test = pd.DataFrame(X_test)
X_test.columns = list(consolidado.drop('song_type', axis=1).columns.values)
y_test = pd.DataFrame(y_test)
y_test.columns = ['song_type']
consolidado_resampled = pd.concat([X_test, y_test], axis=1)[[
    'danceability', 'energy', 'speechiness', 'instrumentalness',
    'valence', 'song_type']]
pred_input_func = tf.estimator.inputs.pandas_input_fn(x=X_test,
                                                     batch_size=10,
                                                     num_epochs=1,
                                                     shuffle=False)
predictions = model.predict(pred_input_func)
pred_list = list(predictions)
nn_pred = []
for pred in pred_list:
    nn_pred.append(pred['class_ids'])
nn_pred = np.asarray(nn_pred)
nn_pred = nn_pred.reshape((-1,))
```

```python
# calcular medidas de error
nn_ac = accuracy_score(y_test, nn_pred)
nn_hl = hamming_loss(y_test, nn_pred)
nn_ll = log_loss(y_test, nn_pred)
print("Accuracy Score: %.2f"%nn_ac)
print("Hamming Loss: %.2f"%nn_hl)
print("Log Loss: %.2f"%nn_ll)
```

```python
# matriz de confusion
confusion_matrix(y_test, nn_pred)
```
